```python
import matplotlib.pyplot as plt
print(plt.__package__)
```

```python
# Creating a simple line chart

# Create a figure with specific size
# width = 10 inches, height = 6 inches
plt.figure(figsize=(10, 6))

# Your plotting code goes here
plt.plot([1, 2, 3], [1, 2, 3])

# Display the plot
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

# Blue line with circle markers
# plt.plot(x, y, 'bo-')
plt.plot(x, y, marker = 'o', linestyle = '-', color = 'purple')

plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Simple Line Plot")
plt.grid(True)
plt.show()
```

```python
# Scatter plot

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

# Red asterisk markers
plt.scatter(x, y, color='teal', marker='*')
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Simple Scatter Plot")
plt.show()
```

```python
# Bar Chart
categories = ['A', 'B', 'C', 'D']  
values = [5, 7, 3, 8]
plt.bar(categories, values, color='lavender')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Simple Bar Chart")
plt.show()
```

```python
# Histogram
import numpy as np
# Generate 1000 random numbers
data = np.random.randn(1000)
plt.hist(data, bins = 30, color='pink', edgecolor='black')
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
# Pie Chart
sizes = [20, 30, 25, 25]
labels = ['Python', 'Java', 'C++', 'Ruby']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']      
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.show()
```
