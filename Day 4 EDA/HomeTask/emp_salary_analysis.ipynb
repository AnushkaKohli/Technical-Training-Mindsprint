```python
import pandas as pd

# Load the dataset
df = pd.read_csv('EMP_DATA.csv')
# Display basic information about the dataset
print(df.info())
```

```python
df
```

```python
# Average Salary in Each Department
average_salary_by_department = df.groupby('department')['salary'].mean()
print("Average Salary in Each Department:")
print(average_salary_by_department)
```

```python
# Highest and Lowest Salary in the Company
highest_salary = df['salary'].max()
lowest_salary = df['salary'].min()
print("Highest Salary in the Company:", highest_salary)
print("Lowest Salary in the Company:", lowest_salary)
```

```python
# Employee with the Maximum Salary
employee_with_max_salary = df[df['salary'] == highest_salary]
print("Employee with the Maximum Salary:")
print(employee_with_max_salary)
```

```python
# Count the Number of Employees in Each Department
employee_count_by_department = df['department'].value_counts()
print("Number of Employees in Each Department:")
print(employee_count_by_department)
```

```python
# Bar Chart Showing the Number of Employees in Each Department
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
employee_count_by_department.plot(kind='bar', title='Number of Employees in Each Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# Plot a Pie Chart Showing Salary Distribution by Department
total_salary_by_department = df.groupby('department')['salary'].sum()
plt.figure(figsize=(10, 6))
total_salary_by_department.plot(kind='pie', title='Salary Distribution by Department', autopct='%1.1f%%')
plt.show()
```

```python
# Scatter Plot Showing Salary vs Experience
plt.figure(figsize=(10, 6))
plt.scatter(df['experience'], df['salary'])
plt.title('Salary vs Experience')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Employees Who Earn More Than the Average Salary
average_salary = df['salary'].mean()
employees_above_average_salary = df[df['salary'] > average_salary]
print("Employees Who Earn More Than the Average Salary:")
print(employees_above_average_salary)
```

```python
# Sort Employees Based on Salary in Descending Order
sorted_employees_by_salary = df.sort_values(by='salary', ascending=False)
print("Employees Sorted by Salary in Descending Order:")
print(sorted_employees_by_salary)
```
