```python
import pandas as pd

# Create student data frame by reading CSV file
df = pd.read_csv('Marks_Data.csv')
df.head()
```

```python
# Check NA values
df.isna()
# Check Null values
df.isnull()
# Fills missing data if any
df.fillna({'Total': 0}, inplace=True)
```

```python
# Calculate total marks of each student 
# axis = 1 means sum should be calculated horizontally across columns for each row
df['Total'] = df[['Math', 'Science', 'English']].sum(axis=1)
df.head()
```

```python
# Function to assign grades
def assign_grade(total):
    if total >= 250:
        return 'A'
    elif total >= 200:
        return 'B'
    elif total >= 150:
        return 'C'
    else:
        return 'D'

# Apply function to assign grades in grade column  
df['Grade'] = df['Total'].apply(assign_grade)

# Display with grades
df[['Name', 'Total', 'Grade']]
```

```python
# Filter students with grade A
grade_a_students = df[df['Grade'] == 'A']
print(grade_a_students)
df.to_csv('Grade_A_Students.csv', index=False)
```

```python
grade_b_students = df[df['Grade'] == 'B']
print("No. of students that secured grade B:", len(grade_b_students))
```
