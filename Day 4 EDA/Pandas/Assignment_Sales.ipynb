```python
import pandas as pd

# Load data into pandas
data = pd.read_csv('Sales_Data.csv')
data
```

```python
# Fill missing data and clean the dataset
```

```python
# Total amount for each order
data['Total_Amount'] = data['Quantity'] * data['Price']
data
```

```python
# Most sold product
# Group by product and sum the quantity
most_sold = data.groupby('Product').sum()['Quantity']
# Get the product with the highest quantity
most_sold = most_sold.idxmax()
print("Most sold product:", most_sold)
```

```python
# Total sales per category
total_sales = data.groupby('Category').sum()['Total_Amount']
print("Total sales per category:")
print(total_sales)
```

```python
# Orders above 10000
orders_above_10000 = data[data['Total_Amount'] > 10000]
print("Orders above 10000:")
print(orders_above_10000)

# Save the processed data into a new CSV file
data.to_csv('Processed_Sales_Data.csv', index=False)
```
